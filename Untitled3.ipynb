```python
def insert_sum(order_book_data,data_to_insert,ob_position):
    top = order_book_data[0:(ob_position-1)]
    bottom = order_book_data[ob_position:]
    QuantityDifference_1 = order_book_data[(ob_position-1):ob_position].QuantityDifference
    QuantityDifference_2 = data_to_insert.QuantityDifference
    order_book_data[(ob_position-1):ob_position][["QuantityDifference"]] = QuantityDifference_1 + QuantityDifference_2
    return pd.concat((top,order_book_data[(ob_position-1):(ob_position)],bottom)).reset_index(drop = True)
```
